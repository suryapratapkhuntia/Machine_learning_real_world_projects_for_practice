```python
from sklearn.preprocessing import OneHotEncoder,OrdinalEncoder
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline

from sklearn.preprocessing import StandardScaler
```

```python
df_cleaned.describe()
```

```python
std_col=df_cleaned[['Area','Production','Yield']]
onehe_col =df_cleaned[['Crop','State','District']]
ordinar_col=df_cleaned[['Season','Production Units']]
```

```python
std=StandardScaler()
std_col_scaled=std.fit_transform(std_col)
std_col_scaled
```

```python
ohe=OneHotEncoder()
onehe_col_encoded=ohe.fit_transform(onehe_col).toarray()
onehe_col_encoded
```

```python
ordinal=OrdinalEncoder()
ordinar_col_encoded=ordinal.fit_transform(ordinar_col)
ordinar_col_encoded
```

```python


std_col=df_cleaned[['Area','Production']]
onehe_col =df_cleaned[['Crop','State','District']]
ordinar_col=df_cleaned[['Season','Production Units']]

preprocessor = ColumnTransformer(
    transformers=[
        ('num', StandardScaler(), std_col.columns),
        ('cat_ohe', OneHotEncoder(handle_unknown='ignore'), onehe_col.columns),
        ('cat_ord', OrdinalEncoder(), ordinar_col.columns)
    ],
    remainder='drop'
)

df_processed = preprocessor.fit_transform(df_cleaned)
display(df_processed)
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import  r2_score
```

```python
x_train,x_test,y_train,y_test=train_test_split(df_processed,df_cleaned['Yield'],test_size=0.2,random_state=42)
```

```python
lin_reg=LinearRegression()
lin_reg.fit(x_train,y_train)
```

```python
y_pred=lin_reg.predict(x_test)
r2_score(y_test,y_pred)
```

```python
from sklearn.tree import DecisionTreeRegressor
```

```python
dt_reg=DecisionTreeRegressor()
dt_reg.fit(x_train,y_train)
```

```python
y_pred=dt_reg.predict(x_test)
r2_score(y_test,y_pred)
```
