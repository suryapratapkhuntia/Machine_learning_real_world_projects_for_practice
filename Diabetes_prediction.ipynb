```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv('/content/diabetes (1).csv')
df.sample(5)
```

```python
print(f"There are {len(df[df['Insulin'] == 0])} rows with 0 in the 'Insulin' column.")
```

```python
df.isnull().sum()
```

```python
df['Insulin'] = df['Insulin'].replace(0, np.nan)
display(df.sample(5))
```

```python
df.isnull().sum()
```

```python
df.describe()
```

```python
df['Insulin'].value_counts()
```

```python
from sklearn.experimental import enable_iterative_imputer
from sklearn.impute import IterativeImputer
```

```python
iter=IterativeImputer()
df['Insulin']=iter.fit_transform(df[['Insulin']])
```

```python
df.describe()
```

```python
from sklearn.preprocessing  import StandardScaler
```

```python
sc=StandardScaler()
df['Insulin']=sc.fit_transform(df[['Insulin']])
```

```python
df['Glucose']=sc.fit_transform(df[['Glucose']])
df['BloodPressure']=sc.fit_transform(df[['BloodPressure']])
df['SkinThickness']=sc.fit_transform(df[['SkinThickness']])
df['Age']=sc.fit_transform(df[['Age']])
```

```python
x=df.drop('Outcome',axis=1)
y=df['Outcome']
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
x_train.shape,x_test.shape,y_train.shape,y_test.shape
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score,confusion_matrix,classification_report
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.svm import SVC
from sklearn.neighbors import KNeighborsClassifier
```

```python
model={
    'LogisticRegression':LogisticRegression(),
    'DecisionTreeClassifier':DecisionTreeClassifier(),
    'RandomForestClassifier':RandomForestClassifier(),
    'SVC':SVC(),
    'KNeighborsClassifier':KNeighborsClassifier()

}

for name,mdl in model.items():
    mdl.fit(x_train,y_train)
    y_pred=mdl.predict(x_test)
    print(name,accuracy_score(y_test,y_pred))
```

```python
dec_clf=DecisionTreeClassifier()
dec_clf.fit(x_train,y_train)
y_pred=dec_clf.predict(x_test)
print(accuracy_score(y_test,y_pred))
```

```python
from sklearn.ensemble import VotingClassifier,BaggingClassifier
```

```python
bagg_clf=BaggingClassifier(estimator=dec_clf,n_estimators=100,random_state=42)
bagg_clf.fit(x_train,y_train)
y_pred=bagg_clf.predict(x_test)
print(accuracy_score(y_test,y_pred))
```

```python
vot_clf=VotingClassifier(estimators=[('lr',LogisticRegression()),('dt',DecisionTreeClassifier())],voting='hard')
vot_clf.fit(x_train,y_train)
y_pred=vot_clf.predict(x_test)
print(accuracy_score(y_test,y_pred))
```
