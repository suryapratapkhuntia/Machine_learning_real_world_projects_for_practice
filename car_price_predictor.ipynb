```python
car=car[car['year'].str.isnumeric()]
```

```python
car['year']=car['year'].astype(np.integer)
```

```python
car['Price']
```

```python

car['Price'] = car['Price'].astype(str)


car['Price'] = car['Price'].replace('Ask For Price', np.nan)


car['Price'] = car['Price'].str.replace(',', '')
```

```python

car['Price'] = pd.to_numeric(car['Price'], errors='coerce')


car['Price'] = car['Price'].astype(float).astype(pd.Int64Dtype())
display(car['Price'].head())


display(car.info())
```

```python

car['kms_driven'] = car['kms_driven'].astype(str).fillna('')


car['kms_driven'] = car['kms_driven'].str.replace(' kms', '')


car['kms_driven'] = car['kms_driven'].str.replace(',', '')


car['kms_driven'] = pd.to_numeric(car['kms_driven'], errors='coerce')
```

```python

car.dropna(subset=['kms_driven'], inplace=True)

car['kms_driven'] = car['kms_driven'].astype(float).astype(pd.Int64Dtype())

display(car.info())
display(car.head())
```

```python
# Take the first three words from the 'name' column
car['name'] = car['name'].str.split().str[:3].str.join(' ')

display(car['name'].head())
```

```python
car.sample(5)
```

```python
car.isnull().sum()
```

```python
from sklearn.experimental import enable_iterative_imputer
from sklearn.impute import IterativeImputer
iter_impute=IterativeImputer()
car['Price']=iter_impute.fit_transform(car[['Price']])
```

```python
car=car[car['Price']<6e6]
```

```python
car
```

```python
# Drop rows with null values in the 'fuel_type' column
car.dropna(subset=['fuel_type'], inplace=True)

# Display the information to show the updated non-null counts
display(car.info())

# Display the null counts to confirm removal
display(car.isnull().sum())
```

```python
x=car.drop(columns='Price')
y=car['Price']
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
print(x_train.shape)
print(x_test.shape)
print(y_train.shape)
print(y_test.shape)
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score
from sklearn.preprocessing import OneHotEncoder
```

```python
ohe=OneHotEncoder(handle_unknown='ignore')
ohe.fit(x[['name','company','fuel_type']])
```

```python
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline, make_pipeline
```

```python
colimn_trans=ColumnTransformer(transformers=[('encoder',ohe,['name','company','fuel_type'])],remainder='passthrough')
```

```python
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor
```

```python
rf_reg=RandomForestRegressor()
```

```python
pipe=make_pipeline(colimn_trans,rf_reg)
```

```python
pipe.fit(x_train,y_train)
```

```python
y_pred=pipe.predict(x_test)
```

```python
r2_score(y_test,y_pred)
```
